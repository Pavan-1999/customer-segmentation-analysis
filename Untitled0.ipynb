```python
import requests
import pandas as pd

def extract_crypto_data():
    url = "https://api.coingecko.com/api/v3/coins/markets"
    params = {
        'vs_currency': 'usd',
        'order': 'market_cap_desc',
        'per_page': 10,
        'page': 1,
        'sparkline': 'false'
    }

    response = requests.get(url, params=params)
    data = response.json()
    df = pd.DataFrame(data)
    return df

# Run extract
crypto_df = extract_crypto_data()
crypto_df[['id', 'symbol', 'current_price', 'market_cap', 'total_volume']].head()
```

```python
def transform_crypto_data(df):
    df_clean = df[['id', 'symbol', 'current_price', 'market_cap', 'total_volume']]
    df_clean.rename(columns={
        'id': 'Name',
        'symbol': 'Symbol',
        'current_price': 'Price_USD',
        'market_cap': 'MarketCap',
        'total_volume': 'Volume'
    }, inplace=True)
    df_clean['Extracted_At'] = pd.Timestamp.now()
    return df_clean

# Run transform
crypto_clean = transform_crypto_data(crypto_df)
crypto_clean.head()
```

```python
import sqlite3

def load_to_sqlite(df, db_name='crypto_data.db'):
    conn = sqlite3.connect(db_name)
    df.to_sql('crypto_prices', conn, if_exists='append', index=False)
    conn.close()
    print("Data loaded successfully.")

# Run load
load_to_sqlite(crypto_clean)
```

```python
# Reconnect and query top 5 cryptos
conn = sqlite3.connect('crypto_data.db')
query = "SELECT * FROM crypto_prices ORDER BY MarketCap DESC LIMIT 5"
top_crypto = pd.read_sql(query, conn)
conn.close()

top_crypto
```
